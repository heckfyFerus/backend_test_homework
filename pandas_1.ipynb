```python
class_23 = titanic[titanic["Pclass"].isin([2, 3])]
class_23
```

**Вышеприведенное эквивалентно фильтрации по строкам, для которых класс равен 2 или 3, и объединению двух операторов с оператором |(или)**

```python
class_23 = titanic[(titanic["Pclass"] == 2) | (titanic["Pclass"] == 3)]
```

**При объединении нескольких условных операторов каждое условие должно быть заключено в круглые скобки ().  Использовать или-|, и-&.**


```python
age_no_na = titanic[titanic["Age"].notna()] #Я хочу работать с данными о пассажирах, возраст которых известен.
age_no_na
```

Меня интересуют имена пассажиров старше 35 лет:

```python
titanic.loc[titanic['Age']>60,'Name']
```

В этом случае подмножество как строк, так и столбцов создается за один раз, и простого использования скобок выбора []уже недостаточно. Операторы loc/ iloc необходимы перед скобками выбора []. При использовании loc/ iloc часть перед запятой — это строки, которые вы хотите, а часть после запятой — это столбцы, которые вы хотите выбрать.

При использовании имен столбцов, меток строк или выражения условия используйте loc оператор перед скобками выбора []. Для части до и после запятой вы можете использовать одну метку, список меток, часть меток, условное выражение или двоеточие. Использование двоеточия указывает, что вы хотите выбрать все строки или столбцы.

```python
titanic.iloc[9:25, 2:5] #Меня интересуют строки с 10 по 25 и столбцы с 3 по 5.
```

```python
titanic.iloc[0:3, 3] = "anonymous" #присвоить имя anonymous первым 3 элементам третьего столбца
```

```python
titanic.head(3) 
```
